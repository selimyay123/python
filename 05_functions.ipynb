```python
# Functions are reusable pieces of code that run when you call them.

# input() function (built-in)
name = input('What is your name?')
print('Hello', name)
```

```python
# int()
# converts a number, boolean, and a numeric string into an integer:
print(int(3.14))
print(int('42'))
print(int(True))
print(int(False))
```

```python
# writing a  custom function
def hello():
    print('hello world')

hello() # calling the function
```

```python
# You can see that our function, calculate_sum, has a and b in its parentheses, separated by a comma. Those are called parameters. Think of parameters as placeholder variables that act as "slots" for the values you pass into functions when you call them.
def calculate_sum(a, b):
    print(a + b)

# To use the parameters, you have to pass in "arguments". Arguments are the values you pass to a function when you call it.
my_sum = calculate_sum(3, 1)
print(my_sum)

# NOTE: If you call the function without the correct number of arguments, you'll get a TypeError

# NOTE: Functions also use a special return keyword to exit the function and return a value. If you don't explicitly use return, Python will return None by default.

```

```python
def calculate_sum(a, b):
    return a + b

my_sum = calculate_sum(3, 1)
print(my_sum)
```
